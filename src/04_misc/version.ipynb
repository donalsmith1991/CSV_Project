```python
import numpy as np

a = np.array([(1, 2, 3), (4, 5, 6)])

# Find the max of the entire array (Default)
print(a.max())         # Output: 6

# Find the max going DOWN the columns
print(a.max(axis=0))   # Output: [4 5 6]

# Find the max going ACROSS the rows
print(a.max(axis=1))   # Output: [3 6]
```
